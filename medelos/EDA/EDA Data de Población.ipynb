```python
# importamos las librerias
import pandas as pd 
import os #Permite acceder a funcionalidades dependientes del SO, en este caso listar contenidos de una carpeta
import re #Especifica un conjunto de cadenas que coinciden con ella

#Se procede a extraer los nombres de cada archivo dataset.
file = [f for f in os.listdir() if re.search('.txt$', f)]
file
```

```python
#Se revisa desde donde empiezan los dataset
file[0]
```

```python
#En la variable frame, se guarda la dataset del primer archivo, usando sus nombres
frame = pd.read_csv(file[0],sep=',', low_memory=False)

# Se agrega una nueva columna  AÑO, que contiene el año en string
frame['AÑO'] = int(file[0][-8:-4])

#En este bucle se guardan los dataset del segundo a ultimo valor, en este caso dentro de frame
for i in file[1:7]:
  aux = pd.read_table(i,sep=',', low_memory=False)
  aux['AÑO'] = i[-8:-4]
  #Añade al dataframe
  frame = pd.concat([aux,frame])
    
#Imprime el frame
print(frame)
```

## 1. ¿Cuántos varones no residentes fueron identificados desde 2018 hasta 2019? 

```python
#Se cargamos en la variable df, los atributos del dataset que se van a usar
df = frame[['RESIDENCIA','SEXO','AÑO','CANTIDAD']]
df
```

```python
#Se usa un filtro para tener solo las personas de residencia extranjera
df = df[df['RESIDENCIA']=='Extranjero']
```

```python
#filtro para tener solo varones
df_temp = df[df['SEXO']=='Hombre']
print(df_temp)
```

```python
#filtro para los años 2018 y 2019
df_temp_2018 = df_temp[df_temp['AÑO']== '2018']
df_temp_2019 = df_temp[df_temp['AÑO']== '2019']
print(df_temp_2019)
```

```python
#Se concatenan los dos filtros en uno solo
df_temp_2018_2019 = pd.concat([df_temp_2018, df_temp_2019])
print(df_temp_2018_2019)
```

```python
#Se cambia el tipo de dato, debido a que su dataset origen era de puros caracteres.
df_temp_2018_2019['CANTIDAD']=df_temp_2018_2019['CANTIDAD'].astype(int)
resultado=sum(df_temp_2018_2019['CANTIDAD'])
print("Número de varones no residentes identificados del 2018 al 2019 son: ", resultado)
```

## 2. Del 2016 al 2018 y en el departamento de Arequipa. ¿Qué porcentaje de mujeres representan las mujeres adolescentes? [Sugerencia: un adolescente es una persona cuya edad está entre 10 a 19 años].

```python
#Se cambian los datos a enteros, como se explicó anteriormente
frame['AÑO'] = frame['AÑO'].astype(int)
frame['EDAD'] = frame['EDAD'].astype(int)
frame['CANTIDAD'] = frame['CANTIDAD'].astype(int)
```

```python
#Se escogen los atributos a utilizar
df1 = frame[['DEPARTAMENTO','SEXO','AÑO','EDAD','CANTIDAD']]
df1
```

```python
#Se realizan los filtros
df1_areq = df1[df1['DEPARTAMENTO']=='Arequipa']
df1_areq_mujeres = df1_areq[df1_areq['SEXO']=='Mujer']

df1_areq_mujeres = df1_areq_mujeres[df1_areq_mujeres['AÑO'] <= 2018]
df1_areq_mujeres = df1_areq_mujeres[df1_areq_mujeres['AÑO'] >= 2016]

df1_areq_joven_mujeres = df1_areq_mujeres[df1_areq_mujeres['EDAD']>=10]
df1_areq_joven_mujeres = df1_areq_joven_mujeres[df1_areq_joven_mujeres['EDAD']<=19]

print(df1_areq_joven_mujeres)
```

```python
#Se suman la cantidad de mujeres de arequipa
mujeres_areq = sum(df1_areq_mujeres['CANTIDAD'])
print('Total de mujeres en Arequipa:', mujeres_areq)
```

```python
#Se suman la cant. de mujeres jóvenes de arequipa
adolescentes_areq = sum(df1_areq_joven_mujeres['CANTIDAD'])

print('Número de adolescentes:', adolescentes_areq)

print(f"El % de mujeres adolescentes en Arequipa en relación a cantidad de mujeres: {adolescentes_areq/mujeres_areq*100}%")
```

## 3. ¿Del 2014 al 2020, en qué año el promedio de edad de las personas mayores (edad igual o mayor de 60 años) identificada fue la más alta?

```python
import warnings as w
w.filterwarnings('ignore')
```

```python
df2 = frame[['AÑO','EDAD','CANTIDAD']]
df2
```

```python
df2_adulto_mayor = df2[df2['EDAD'] >= 60]
df2_adulto_mayor
```

```python
#se crea un nuevo atributo multiplicando los atributos edad y cantidad
df2_adulto_mayor['EDAD_POR_CANTIDAD'] = df2_adulto_mayor['EDAD'] * df2_adulto_mayor['CANTIDAD']
print(df2_adulto_mayor)
```

```python
#Se agrupa la info por año.
adultos_grupo = df2_adulto_mayor.groupby(['AÑO'],as_index=False).sum()
print(adultos_grupo)
```

```python
#se crea un nuevo atributo dividiento edad por cantidad, entre cantidad
adultos_grupo['PROM_EDAD'] = adultos_grupo['EDAD_POR_CANTIDAD']/adultos_grupo['CANTIDAD']
adultos_grupo
```

```python
#Se ordena de acuerdo al mayor promedio de edad
adultos_grupo.sort_values(by=['PROM_EDAD'],inplace=True, ascending=False)
adultos_grupo[['AÑO','PROM_EDAD']]
```

```python
print("El mayor promedio de edad de personas adultas mayores a 60 años es del 2018 con ", adultos_grupo['PROM_EDAD'].max())
```

## 4.  Indique las 5 ciudades con mayor edad media, tanto de hombres como de mujeres

```python
#Se reutiliza el procedimiento anterior, usando los siguientes datos
df3 = frame[['SEXO','EDAD','CANTIDAD','PROVINCIA']]
df3
```

```python
# realizamos los filtros para los Hombres
hombres = df3[df3['SEXO'] == 'Hombre']
hombres_grupo=hombres.groupby(["PROVINCIA"],as_index=False).mean()
hombres_grupo.sort_values(by=["EDAD"],inplace=True, ascending=False)
hombres_grupo.head(5)
```

```python
#Realizamos los filtros para las Mujeres
mujeres = df3[df3['SEXO'] == 'Mujer']
mujeres_grupo = mujeres.groupby(['PROVINCIA'],as_index=False).mean()
mujeres_grupo.sort_values(by=["EDAD"],inplace=True, ascending=False)
mujeres_grupo.head(5)
```

## 5. Por favor, compruebe si las edades medias de mujeres y hombres son estadísticamente diferentes
#### Sugerencia: Puedes trazar un histograma para explicar tu idea.

```python
#Escogemos los atributos a utilizar
df4 = frame[['SEXO','EDAD', 'AÑO', 'CANTIDAD']]

df4 ['EDAD_POR_CANTIDAD'] = df4 ['EDAD'] * df4 ['CANTIDAD']

df4_mujeres=df4[df4["SEXO"]=="Mujer"]
df4_hombres=df4[df4["SEXO"]=="Hombre"]

cant_total=df4['CANTIDAD'].sum()

promEdades_hombres=df4_hombres['EDAD_POR_CANTIDAD'].sum()/cant_total
promEdades_mujeres=df4_mujeres['EDAD_POR_CANTIDAD'].sum()/cant_total

sexo=['Hombre','Mujer']
lista_prom=[promEdades_hombres,promEdades_mujeres]

df4_promedios_sexo=pd.DataFrame(list(zip(sexo,lista_prom)),columns=['SEXO','EDAD PROM'])
df4_promedios_sexo
```
